```python
with open('auxilio_a_pesquisadores_-_fapeg_2013.csv', newline='') as infile:
    reader = csv.reader(infile, delimiter=',')
    next(reader)
    rows = [row for row in reader if reader.line_num not in range(963, 967)]
    rows = sorted(rows, key=lambda x: float(x[3].replace(',', '.')), reverse=True)

with open('auxilio_a_pesquisadores_modfied.csv', 'w', newline='') as outfile:
    writer = csv.writer(outfile)
    writer.writerow(['Credor Fornecedor', 'Valor Empenhado', 'Valor Pago', 'Valor a Pagar'])
    for row in rows:
        writer.writerow([row[0], row[3], row[4], row[5]])
```

```python
with open('auxilio_a_pesquisadores_modfied.csv', newline='') as infile:
    reader = csv.reader(infile, delimiter=',')
    next(reader)
    data = []
    for row in reader:
        if reader.line_num in range(963, 967):
            continue
        row[1] = float(row[1].replace(',', '.'))
        row[2] = float(row[2].replace(',', '.'))
        row[3] = float(row[3].replace(',', '.'))
        data.append(row)
    data.sort(key=lambda x: x[1], reverse=True)
```

```python
with open('auxilio_a_pesquisadores_modfied.csv', newline='') as infile, \
        open('auxilio_a_pesquisadores_filtered.csv', 'w', newline='') as outfile:
    reader = csv.reader(infile, delimiter=',')
    writer = csv.writer(outfile, delimiter=',')
    writer.writerow(next(reader))
    for row in reader:
        if row[1] == '0' and row[2] == '0' and row[3] == '0':
            continue
        writer.writerow(row)
```

```python
credor_fornecedor = []
valor_empenhado = []
valor_pago = []
valor_a_pagar = []

with open('auxilio_a_pesquisadores_modfied.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    next(reader) # pula o cabeçalho
    for row in reader:
        credor_fornecedor.append(row[0])
        valor_empenhado.append(float(row[1].replace(',', '.')))
        valor_pago.append(float(row[2].replace(',', '.')))
        valor_a_pagar.append(float(row[3].replace(',', '.')))
print("------------------------------------------------")
print("Total Valor Empenhado: R$", sum(valor_empenhado))
print("Média Valor Empenhado: R$", np.mean(valor_empenhado))
print("Desvio Padrão Valor Empenhado: R$", np.std(valor_empenhado))
print("Mediana Valor Empenhado: R$", np.median(valor_empenhado))
print("------------------------------------------------")
print("Total Valor Pago: R$", sum(valor_pago))
print("Média Valor Pago: R$", np.mean(valor_pago))
print("Desvio Padrão Valor Pago: R$", np.std(valor_pago))
print("Mediana Valor Pago: R$", np.median(valor_pago))
print("------------------------------------------------")
print("Total Valor a Pagar: R$", sum(valor_a_pagar))
print("Média Valor a Pagar: R$", np.mean(valor_a_pagar))
print("Desvio Padrão Valor a Pagar: R$", np.std(valor_a_pagar))
print("Mediana Valor a Pagar: R$", np.median(valor_a_pagar))
print("------------------------------------------------")

#------------------------------------------------
#Total Valor Empenhado: R$ 22.603.271,64
#Média Valor Empenhado: R$ 23.520,57
#Desvio Padrão Valor Empenhado: R$ 279.357,63
#Mediana Valor Empenhado: R$ 8.400,0
#------------------------------------------------
#Total Valor Pago: R$ 9.367.259,64
#Média Valor Pago: R$ 9.747,40
#Desvio Padrão Valor Pago: R$ 6.311,16
#Mediana Valor Pago: R$ 8.400,0
#------------------------------------------------
#Total Valor a Pagar: R$ 13.236.012,0
#Média Valor a Pagar: R$ 13.773,16
#Desvio Padrão Valor a Pagar: R$ 279.762,91
#Mediana Valor a Pagar: R$ 0.0
#------------------------------------------------

# Observando os dados, o que se pode observar, é que a média e mediana não demonstram um valor próximo
# a verdadeira média, por isso a importancia do desvio padrão. 
# Um baixo desvio padrão indica que os pontos dos dados tendem a estar próximos da média ou do valor esperado
# Um alto desvio padrão indica que os pontos dos dados estão espalhados por uma ampla gama de valores
# Como mostra os dados acima, o desvio padrão é baixo.
```
