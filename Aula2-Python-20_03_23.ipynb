```python
# Aprendendo a abrir arquivo, escrever tabelas no arquivo
import csv
import collections

brasilCsv = open('brasil.csv', encoding='utf-8')
populacao = collections.Counter()

for eachRow in csv.DictReader(brasilCsv):
  populacao[eachRow['estado']] += int(eachRow['habitantes'])

file_csv = open("estados.csv", mode="w", encoding="utf-8")

with open("estados.csv", "w") as newFile:
  newFile = csv.writer(newFile, delimiter=",")
  newFile.writerow(["estado", "população"])

  for estado,populacao in populacao.most_common():
    newFile.writerow([estado,populacao])

estadosCsv = open('estados.csv', encoding='utf-8')
for eachRow in csv.DictReader(estadosCsv):
  print(eachRow)
```

```python
# Atividade: Total de habitantes do sudeste
import csv

sudeste = ["SP", "RJ", "ES", "MG"]
total_habitantes = 0

with open("brasil.csv", "r", encoding="utf-8") as arquivo:
    leitor = csv.DictReader(arquivo)
    for eachRow in leitor:
        estado = eachRow["estado"]
        habitantes = int(eachRow["habitantes"])
        if estado in sudeste:
            total_habitantes += habitantes

print("Total de habitantes do Sudeste:", total_habitantes)
```

```python
# Atividade: média de habitantes por área por estados
import csv
from collections import defaultdict

estados = defaultdict(lambda: [0, 0])

with open('brasil.csv', 'r', encoding='utf-8') as f:
    reader = csv.reader(f)
    next(reader)
    for eachRow in reader:
        estado, _, habitantes, area = eachRow
        estados[estado][0] += float(area)
        estados[estado][1] += int(habitantes)

media_habitantes_por_area = {}
for estado, (area, habitantes) in estados.items():
    media_habitantes_por_area[estado] = habitantes / area

for estado, media in media_habitantes_por_area.items():
    print(f'{estado}: {media:.2f} habitantes por km²')
```
