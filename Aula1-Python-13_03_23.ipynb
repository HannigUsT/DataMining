```python
from urllib.request import urlopen

world = input('Digite uma palavra para verificar se a mesma existe no dicionário: ')

with urlopen("https://www.ime.usp.br/~pf/dicios/br-sem-acentos.txt") as response:
   body = response.read().decode('utf-8')

if (body.find(world) >= 0):
    print('Existe')
else:
  print('Não existe')
```
